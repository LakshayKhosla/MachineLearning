```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import bokeh as bk
import numpy as np
```

```python
df_Petrol = pd.read_excel("./Dataset-Petroleum.xlsx")
```

```python
df_Petrol.head()
```

### Petrol Prices - Proceeding with Visual Analysis

```python
df_Petrol = df_Petrol.set_index("Date")
```

```python
df_Petrol.plot(figsize=(20,10))
plt.xlabel("Year", size = 24)
plt.ylabel("Price / Litre (INR)", size = 24)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.title("Petrol Price in 4 states (1989 - 2018) - Analysis", size = 30)
```

## Bifurcating 4 states in Time Series Plot

```python
sns.set()
plt.figure(figsize=(20,10))
sns.lineplot(data=df_Petrol, style="choice", markers=True, hue="States")
plt.xlabel("Year", size = 24)
plt.ylabel("Price / Litre (INR)", size = 24)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.title("Petrol Price in 4 states (1989 - 2018) - Analysis", size = 30)
```

## Highlighting the data points
### From 1990 to 2018, Petrol Prices went upto 9.5 times
### Minimum 8.5 Rupee / Litre in 1989 -- 78.43 Rupee / Litre in 2018

```python
with sns.axes_style('white'):
    g1 = sns.jointplot(x="Delhi", y="Kolkata", data=df_Petrol, kind="reg")
    fig1 = g1.ax_joint.plot(np.linspace(0, 100),
                    np.linspace(25, 100), ':k')
    plt.xlabel("Delhi", size = 24)
    plt.ylabel("Kolkata", size = 24)
    plt.xticks(size = 20)
    plt.yticks(size = 20)
```

### Delhi v/s Kolkata

```python
      
    
with sns.axes_style('white'):
    g2 = sns.jointplot(x="Delhi", y="Mumbai", data=df_Petrol, kind="reg")
    fig2 = g2.ax_joint.plot(np.linspace(0, 100),
                    np.linspace(25, 100), ':k')
    plt.xlabel("Delhi", size = 24)
    plt.ylabel("Mumbai", size = 24)
    plt.xticks(size = 20)
    plt.yticks(size = 20)
```
